# 05_04 - CNN-BiLSTM Sequence (index)

CNN-BiLSTM training was split into **three notebooks** (one seed each) because a single notebook with all seeds exceeds practical Colab runtime.

## Run order

1. `05_04a_CNN_BiLSTM_Sequence_seed42.ipynb` — canonical SSOT paths (required for headline)
2. `05_04b_CNN_BiLSTM_Sequence_seed123.ipynb` — optional multi-seed stability
3. `05_04c_CNN_BiLSTM_Sequence_seed456.ipynb` — optional multi-seed stability

Then run `05_05_Weighted_Blending.ipynb` (loops all `SEEDS` and reads matching `05_01`/`05_02`/`05_04` artifacts per seed).
